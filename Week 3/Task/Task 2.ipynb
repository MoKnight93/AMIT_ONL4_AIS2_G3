
thanos project. 
make a function to do the following:. 
1-open the whole folder. 
2-create number of files inside it. 
3-check the len of files in that folder. 
4-delete random half of this folder. 
5- check the result. 

```python
import os
import random

def thanos_project():
    """
    Thanos project: Delete random half of files in a folder
    """
    
    # Step 1: Open the whole folder
    folder_name = "thanos_folder"
    
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    print("Step 1: Folder opened/created")
    
    # Step 2: Create number of files inside it
    num_files = 10
    
    for i in range(1, num_files + 1):
        with open(f"{folder_name}/file_{i}.txt", 'w') as f:
            f.write(f"File {i}")
    
    print(f"Step 2: Created {num_files} files")
    
    # Step 3: Check the len of files in that folder
    files = os.listdir(folder_name)
    print(f"Step 3: Total files = {len(files)}")
    
    # Step 4: Delete random half of this folder
    half = len(files) // 2
    files_to_delete = random.sample(files, half)
    
    for file in files_to_delete:
        os.remove(f"{folder_name}/{file}")
    
    print(f"Step 4: Deleted {half} files")
    
    # Step 5: Check the result
    remaining_files = os.listdir(folder_name)
    print(f"Step 5: Remaining files = {len(remaining_files)}")


# Run the function
thanos_project()
```
